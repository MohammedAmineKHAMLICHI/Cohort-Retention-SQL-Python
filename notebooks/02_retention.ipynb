# Cohort retention (Python)


```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ROOT = Path('..').resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.retention import build_retention

orders = pd.read_csv(ROOT / 'data' / 'orders.csv', parse_dates=['order_date'])
ret = build_retention(orders)
ret
```

```python
value_columns = [c for c in ret.columns if isinstance(c, (int, np.integer))]
plt.figure(figsize=(8, 4))
plt.imshow(ret[value_columns].values, aspect='auto')
plt.colorbar(label='% active')
plt.yticks(range(len(ret.index)), ret.index)
plt.xticks(range(len(value_columns)), value_columns)
plt.title('Cohort retention heatmap')
plt.xlabel('Months since acquisition')
plt.ylabel('Cohort')
plt.tight_layout()
plt.show()
```
